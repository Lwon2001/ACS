```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# save figures
import os
from pathlib import Path
PROJECT_ROOT_DIR = Path(os.getcwd())
TEST_GROUP = 'changed_label2'
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", TEST_GROUP)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
# Population size
log_1 = np.genfromtxt('log_1.txt', delimiter='\n', skip_header=True)
log_2 = np.genfromtxt('log_2.txt', delimiter='\n', skip_header=True)
log_3 = np.genfromtxt('log_3.txt', delimiter='\n', skip_header=True)
log_4 = np.genfromtxt('log_4.txt', delimiter='\n', skip_header=True)
log_5 = np.genfromtxt('log_5.txt', delimiter='\n', skip_header=True)

print(log_1.size, log_2.size, log_3.size, log_4.size, log_5.size)

ps = pd.DataFrame({'50':log_1, '100':log_2, '200':log_3, '400':log_4, '800':log_5})

plt.figure()
ps.plot.box(figsize=[12,10], fontsize=16, title='Population size')
save_fig('Pop_size')
```

```python
# Mutation rate
log_2_1 = np.genfromtxt('log_2_1.txt', delimiter='\n')
log_2_2 = np.genfromtxt('log_2_2.txt', delimiter='\n')
log_2_3 = np.genfromtxt('log_2_3.txt', delimiter='\n')
log_2_4 = np.genfromtxt('log_2_4.txt', delimiter='\n')

mr = pd.DataFrame({'0.2':log_5, '0.1':log_2_1, '0.05':log_2_2, '0.03':log_2_3, '0.01':log_2_4})

plt.figure()
mr.plot.box(figsize=[12,10], fontsize=16, title='Mutation rate')
save_fig('Mutation_rate')
```

```python
# Crossover rate
log_3_1 = np.genfromtxt('log_3_1.txt', delimiter='\n')
log_3_2 = np.genfromtxt('log_3_2.txt', delimiter='\n')
log_3_3 = np.genfromtxt('log_3_3.txt', delimiter='\n')
log_3_4 = np.genfromtxt('log_3_4.txt', delimiter='\n')

cr = pd.DataFrame({'0.8':log_4, '0.85':log_3_1, '0.9':log_3_2, '0.95':log_3_3, '0.98':log_3_4})

plt.figure()
cr.plot.box(figsize=[12,10], fontsize=16, title='Crossover rate')
save_fig('Crossover_rate')
```

```python
# Tournament size
log_4_1 = np.genfromtxt('log_4_1.txt', delimiter='\n')
log_4_2 = np.genfromtxt('log_4_2.txt', delimiter='\n')
log_4_3 = np.genfromtxt('log_4_3.txt', delimiter='\n')
log_4_4 = np.genfromtxt('log_4_4.txt', delimiter='\n')

cr = pd.DataFrame({'8':log_4_3, '5':log_3_2, '4':log_4_2, '2':log_4_4})

plt.figure()
cr.plot.box(figsize=[12,10], fontsize=16, title='Tournament size')
save_fig('Tournament_size')
```
